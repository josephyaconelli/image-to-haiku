```python
output_file = "./data/haiku_cleaned_data-NEW.txt"
input_file  = "./data/all_haiku.csv"
```

```python
import pandas as pd
```

```python
poetry_df = pd.read_csv(input_file)
poetry_df
```

```python
poetry_df['source'].unique()
```

```python
sballas = poetry_df.loc[poetry_df['source'] == 'sballas'].index
img2poems = poetry_df.loc[poetry_df['source'] == 'img2poems'].index
twaiku = poetry_df.loc[poetry_df['source'] == 'twaiku'].index
```

```python
poetry_df_clean = poetry_df.drop(sballas)
poetry_df_clean = poetry_df_clean.drop(img2poems)
poetry_df_clean = poetry_df_clean.drop(twaiku)
poetry_df_clean
```

```python
training_df = poetry_df_clean[["0", "1", "2"]]
training_df
```

```python
import re

no_alpha_pattern = re.compile('[^a-zA-Z_ ]+')
dash_pattern = re.compile('[-\/]+')
multiple_spaces_pattern = re.compile('[ ]{2,}')
```

```python
output = ""
dash_pattern = re.compile('[-]+')
for i, row in training_df.iterrows():
    output += "startlinehere"
    for col in row:
        lower_line = str(col).lower() # make lower case
        clean_line = dash_pattern.sub(' -- ', lower_line) # replace dashes and slashes with spaces
        #clean_line = no_alpha_pattern.sub('', clean_line) # remove all non-alphabet symbols
        clean_line = re.sub( r'([a-zA-Z])([,.!-])', r'\1 \2', clean_line )
        clean_line = multiple_spaces_pattern.sub(' ', clean_line) # replace multiple spaces with just one

        if len(clean_line) == 0:
            continue
        if clean_line[0] == ' ':
            clean_line = clean_line[1:]
        if len(clean_line) == 0:
            continue
        if clean_line[-1] == ' ':
            clean_line = clean_line[:-1:]
        s = 0
        output += " " + clean_line + " lineendhere"
    output += " endlinehere "
```

```python
with open(output_file, 'w') as f:
    f.write(output)
```
